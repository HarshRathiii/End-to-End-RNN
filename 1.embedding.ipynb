```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'I love programming in Python',
    'Python is a great programming language',
    'I enjoy solving problems with Python',
    'Python is versatile and powerful',
    'I love learning new things in Python',
    'Python is my favorite programming language',
    'I enjoy coding in Python'
]
```

```python
## define the vocabulary size
voc_size = 10000
```

```python
## One hot representation for every word
one_hot_repr = [one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding
#from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_lenght = 8
embedded_docs = pad_sequences(one_hot_repr, maxlen=sent_lenght, padding='pre')
embedded_docs
```

```python
embedded_docs.shape
```

```python
## feature representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_lenght))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
